```python
import math

def add(x, y):
    return x + y

def multiply(a, b):
    return a * b
```

```python
class DataProcessor:
    def __init__(self, name):
        self.name = name

    def process(self, data):
        result = add(data, 1)
        return multiply(result, 2)
```

```python
processor = DataProcessor('test')
output = processor.process(5)
print(output)
```
